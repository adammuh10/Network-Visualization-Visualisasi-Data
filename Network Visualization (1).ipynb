```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
G = nx.Graph()

G.add_node('bekasi',role='bekasi')
G.add_node('bogor',role='bogor')
G.add_node('depok',role='depok')
G.add_node('sukabumi',role='sukabumi')
G.add_node('cianjur',role='cianjur')
G.add_node('karawang',role='karawang')
G.add_node('purwakarta',role='purwakarta')
G.add_node('subang',role='subang')
G.add_node('indramayu',role='indramayu')
G.add_node('cirebon',role='cirebon')
G.add_node('kuningan',role='kuningan')
G.add_node('majalengka',role='majalengka')
G.add_node('sumedang',role='sumedang')
G.add_node('bandung',role='bandung')
G.add_node('cimahi',role='cimahi')
G.add_node('garut',role='garut')
G.add_node('tasikmalaya',role='tasikmalaya')
G.add_node('ciamis',role='ciamis')

G.nodes(data=True)
```

```python
G.add_edge('bandung','cimahi',weight=50,relation='1')
G.add_edge('depok','bekasi',weight=7,relation='2')
G.add_edge('bogor','depok',weight=11,relation='3')
G.add_edge('bogor','cianjur',weight=10,relation='4')
G.add_edge('bogor','sukabumi',weight=9,relation='5')
G.add_edge('bekasi','karawang',weight=5,relation='6')
G.add_edge('purwakarta','karawang',weight=9,relation='7')
G.add_edge('purwakarta','subang',weight=17,relation='8')
G.add_edge('cianjur','sukabumi',weight=12,relation='9')
G.add_edge('bandung','cianjur',weight=15,relation='10')
G.add_edge('bandung','sumedang',weight=3,relation='11')
G.add_edge('majalengka','sumedang',weight=4,relation='12')
G.add_edge('ciamis','tasikmalaya',weight=15,relation='13')
G.add_edge('indramayu','cirebon',weight=18,relation='14')
G.add_edge('purwakarta','cimahi',weight=5,relation='15')
G.add_edge('kuningan','cirebon',weight=10,relation='16')
G.add_edge('garut','tasikmalaya',weight=8,relation='17')
G.add_edge('bandung','garut',weight=9,relation='18')
G.add_edge('majalengka','cirebon',weight=12,relation='19')
G.add_edge('subang','cirebon',weight=20,relation='20')
G.add_edge('purwakarta','indramayu',weight=40,relation='21')
G.add_edge('tasikmalaya','sumedang',weight=0,relation='22')

G.edges(data=True)
```

```python
nx.draw_networkx(G, with_labels=True)
```

```python
list(G.edges)
```

```python
list(G.neighbors('purwakarta'))
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

# Kesimpulan
- Dari Graph diatas dapat disimpulkan bahwa Kota Purwakarta, Kota Cirebon dan Kota Bandung memiliki nilai degree centrality tertinggi dengan nilai 0.23529411764705882.
- Lalu untuk nilai betweenness centrality tertinggi juga diraih oleh kota Purwakarta, Cirebon dan Bandung dengan nilai 0.4301470588235294.
- Sehingga dapat disimpulkan bahwa kota purwakarta, cirebon dan bandung merupakan kota terpenting di jawa barat
